```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
from scipy import integrate
```

```python
arr = np.loadtxt("droptower_vdata.txt")
```

```python
time = arr[:,0]
velocity = arr[:,1]
position = scp.integrate.cumtrapz(velocity, time, initial = 0)
acceleration = np.insert(np.diff(velocity), 0, 0)

f, ((ax0, ax1, ax2)) = plt.subplots(3,1)
ax0.plot(time, acceleration, 'red')
ax1.plot(time, velocity, 'blue')
ax2.plot(time, position, 'black')

ax0.set_title('Acceleration')
ax1.set_title('Velocity')
ax2.set_title('Position')
f.tight_layout()
f.suptitle('DROP TOWER BEHAVIOR')
f.subplots_adjust(top = 0.85)
```

```python
pos_acceleration = [ele if ele > 0 else 0 for ele in acceleration]
print(pos_acceleration)
index = [i for i, j in enumerate(pos_acceleration) if j == 0]
pos_acceleration1 = np.delete(pos_acceleration, index)
print(pos_acceleration1)
print "The Average is " + str(np.average(pos_acceleration))
```
